```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Path of the files to read
cancer_filepath = "../input/cancer.csv"

# Fill in the line below to read the file into a variable cancer_data
cancer_data = pd.read_csv(cancer_filepath, index_col = 'Id')
```

```python
cancer_data.head()
```

```python
# Histograms for benign and maligant tumors
# Use this nice way, of saying data=THIS, and then just define features and hue is the classes!
plt.figure(figsize=(8,6))
plt.title("Area of tumors and diagnosis")
sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')
```

#### We observe that the area of belign tumors are way less than the area of maligant tumors. Should test further if area > 500

```python
# We will now do a KDE plot (showing the distributions)
# KDE plots for benign and malignant tumors
plt.figure(figsize=(8,6))
plt.title("Distributions of tumor radius")
sns.kdeplot(data=cancer_data, x='Radius (worst)', hue='Diagnosis', fill=True)
```

#### Again we observe that malignant tumors has larger radius

# Iris 2d KDE plots

```python
# Path of the file to read
iris_filepath = "../input/iris.csv"

# Read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
# Histograms for each species
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species")
```

```python
# KDE plots for each species
plt.figure(figsize=(10,6))
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', fill=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```
