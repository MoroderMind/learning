```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Im choosing a bookstore dataset

```python
path = "../input/bookdata.csv"
df = pd.read_csv(path)
#Changing stars to category, so i can plot better
df['stars'] = df['stars'].astype('category')
```

```python
df.head()
```

```python
df.shape, df.columns
```

### Hmm, what type of catagory has what prices and stars?

```python
plt.figure(figsize=(8,6))
sns.set_style("dark")
sns.regplot(x=df['price'][:50], y=df['stars'][:50])
```

```python
df['category'].nunique()
```

### There are 50 different categories. Nothing seems appropiate
##### Maybe stars and price?

```python
plt.figure(figsize=(8,6))
plt.title("Stars and price?")
sns.kdeplot(data=df, x='price', hue='stars', fill=True)
```

```python
sns.histplot(data=df, x='price', hue='stars', fill=True)
```

```python
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.show()
```

```python
sns.boxplot(x='product_type', y='price', data=df)
plt.title('Boxplot of Prices by Product Type')
plt.show()
```

```python
sns.scatterplot(x='price', y='num_reviews', data=df)
plt.title('Scatter Plot of Price vs. Number of Reviews')
plt.show()
```
